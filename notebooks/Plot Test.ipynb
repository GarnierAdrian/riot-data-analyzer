```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/interm/player_frames.csv')
df[df['summonerName']=='Sun']
```

```python
df.describe()
```

```python
for pnum in df['stats.participantId'].unique():
    if pnum <= 5:
        plt.scatter(df[df['stats.participantId'] == pnum].timestamps//60000,df[df['stats.participantId'] == pnum].minionsKilled,label=pnum)
plt.show()

fig = plt.figure(figsize=[15, 15])
ax = fig.add_subplot(111, projection='3d')
for pnum in df['stats.participantId'].unique():
    ax.scatter(df[df['stats.participantId'] == pnum].timestamps//60000,df[df['stats.participantId'] == pnum].xp,df[df['stats.participantId'] == pnum].minionsKilled,label=pnum)
ax.set_xlabel('Game time (minutes)')
ax.set_ylabel('Level')
ax.set_zlabel('Minions Killed')
ax.legend()
plt.show()
```
