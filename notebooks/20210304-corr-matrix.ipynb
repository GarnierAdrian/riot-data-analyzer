```python
import pandas as pd
data = pd.read_csv("../data/interm/game.csv")
print(data.columns)
```

```python
data['stats.win'] = data['stats.win'].astype(int)
data.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
data.hist(bins=50, figsize=(40,40))
```

```python
import plotly.express as px
championData = data.groupby("championName").mean()
fig1 = px.scatter_matrix(
        championData, dimensions=['stats.win'], color="championName")
fig1.show()
```

```python
corr_matrix = data.corr()
```

```python
corr_matrix['stats.win'].sort_values(ascending=False)
```

```python
corr_matrix['stats.win'].sort_values(ascending=True)
```
